```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data structure: 


```python
# [id, 2021, 2022, 2023, 2024]
sales_data = np.array([
[1, 150000, 180000, 220000, 250000], # Paradise Biryani
[2, 120000, 140000, 160000, 190000], # Beijing Bites
[3, 200000, 230000, 260000, 300000],
# Pizza Hub
[4, 180000, 210000, 240000, 270000], # Burger Point
[5, 160000, 185000, 205000, 230000]
# Chai Point
])

print("===== Zomato Sales analysis ====")
print(f"sales data shape {sales_data.shape}")
print(f"First 3 restorent data = \n{sales_data[:3]}")
print(f"restorent data = \n{sales_data[0:3,1:4]}")
print(f"restorent data = \n{sales_data[:,  :3]}")
print(f"restorent data = \n{sales_data[:,1:]}")
```

# total sale 



```python
yearly_total = np.sum(sales_data[:,1:], axis=0)
print(yearly_total)
```

## minimum sales per restaurent

```python
min_sales = np.min(sales_data[:,1:], axis=1)
print(min_sales)
```

## max sales per year 

```python
max_sales = np.max(sales_data[:,1:], axis=0)
print(max_sales)
```

## avg sales per restaurent

```python
avg_sales = np.mean(sales_data[:,1:], axis=1)
print(avg_sales)
```

```python
cumulative_sum = np.cumsum(sales_data[:,1:] , axis=1)
print(cumulative_sum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumulative_sum, axis=0))
plt.title("Average cumulative sales accross all restaurent ")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
monthly_avg = sales_data[:,1:] / 12
print(monthly_avg)
```
